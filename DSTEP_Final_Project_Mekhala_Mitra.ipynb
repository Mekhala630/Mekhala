**PROBLEM STATEMENT**


The demographic analysis project for St. Lawrence County, New York addresses the need for a comprehensive understanding of the household characteristics within the county. With an increasingly diverse and evolving demographic landscape, local authorities, policymakers, and community stakeholders lack a detailed insight into the distribution of households and family structures demographics. The problem statement underscores the necessity for a nuanced understanding of St. Lawrence County's demographic profile to support informed and strategic actions that align with the evolving needs of its residents.

```python
import numpy as np
import pandas as pd
import geopandas as gp
import matplotlib.pyplot as plt
import contextily as ctx
```

```python
# Load the census data
fname="/content/drive/MyDrive/DSTEP/DSTEP_project_data/DSTEP_household_data/ACSST5Y2018.S1101-Data.csv"
```

```python
census=pd.read_csv(fname,skiprows=1)
```

```python
# Drop columns containing 'Error' in their name from the 'census' DataFrame
columns_to_drop = [col for col in census.columns if 'Error' in col]
census = census.drop(columns=columns_to_drop)
```

```python
census
```

```python
# Remove words followed by '!!' from column names
census.columns = census.columns.str.replace(r'\w+!!', '', regex=True)
# Replace spaces with underscores in column names
census.columns = census.columns.str.replace(' ', '_')
census
```

```python
#Loading Census tract shapes
sname="/content/drive/MyDrive/DSTEP/DSTEP_project_data/tl_2018_36_tract"

ctshp=gp.read_file(sname)
```

```python
ctshp.plot()
```

```python
ctshp
```

```python
# Merge the census data with tract shapes

# Remove '1400000US' prefix from 'Geography' column in 'census'
census['Geography'] = census['Geography'].str.replace('1400000US', '')


# Merge the two DataFrames based on the modified 'GEOID' and 'Geography' columns
merged_df = pd.merge(ctshp, census, left_on="GEOID", right_on='Geography')

merged_df
```

```python
from pyproj import CRS
merged_df = merged_df.to_crs(epsg=4326)
```

```python
#Choropleth Map of Average Household Size in St. Lawrence County

merged_df['Average_household_size'] = pd.to_numeric(merged_df['Average_household_size'], errors='coerce')

fig, ax = plt.subplots(figsize=(9, 7))

# Plot the choropleth map with a continuous color scale
plot = merged_df.plot(column='Average_household_size', cmap='viridis', linewidth=0.8, ax=ax, edgecolor='0.8', legend=False)

# Add a title to the plot
plt.title('Choropleth Map of Average Household Size in St. Lawrence County')

# Add a continuous colorbar to the plot
sm = plt.cm.ScalarMappable(cmap='viridis', norm=plt.Normalize(vmin=merged_df['Average_household_size'].min(), vmax=merged_df['Average_household_size'].max()))
sm._A = []
cbar = plt.colorbar(sm, ax=ax, orientation='vertical', fraction=0.03, pad=0.1)
cbar.set_label('Average Household Size')

# Display the plot
plt.show()
```

***Interpretation***

**The average household size in different regions provides a snapshot of family structures, indicating whether households tend to be larger or smaller.**

```python
#Choropleth Map of Total no of Households in St. Lawrence County

merged_df = merged_df.loc[:, ~merged_df.columns.duplicated()]
fig, ax = plt.subplots(figsize=(9, 7))

# Plot the choropleth map with a continuous color scale
plot = merged_df.plot(column='Total_households', cmap='viridis', linewidth=0.8, ax=ax, edgecolor='0.8', legend=False)

# Add a title to the plot
plt.title('Choropleth Map of total number of Households in St. Lawrence County')

# Add a continuous colorbar to the plot
sm = plt.cm.ScalarMappable(cmap='viridis', norm=plt.Normalize(vmin=merged_df['Total_households'].min(), vmax=merged_df['Total_households'].max()))
sm._A = []
cbar = plt.colorbar(sm, ax=ax, orientation='vertical', fraction=0.03, pad=0.1)
cbar.set_label('Total Households')

# Display the plot
plt.show()
```

***Observations:***

1. **The total number of households in different regions indicates population density, showing areas with higher concentrations of residents.**


2. **It provides a measure of the size and demand for services within each region, helping in strategic planning for infrastructure, education, healthcare etc.**

```python
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt

x_column = 'Average_household_size'
y_column = 'Total_households'

# Extract the data for the two columns
x_data = merged_df[x_column]
y_data = merged_df[y_column]

# Calculate the linear regression
slope, intercept, r_value, p_value, std_err = linregress(x_data, y_data)

# Calculate R2
r_squared = r_value**2

# Plotting the scatter plot with trend line
plt.figure(figsize=(8, 6))  # Adjust the size of the figure (smaller)
plt.scatter(x_data, y_data, label='Data Points')
plt.plot(x_data, slope * x_data + intercept, color='red', label=f'Linear Regression (R-squared={r_squared:.2f})')
plt.xlabel('Average size of household')
plt.ylabel('Total no of household')
plt.title('Scatter Plot with Linear Regression')
plt.legend()

# Show the plot
plt.show()

print(f'R-squared value: {r_squared}')

plt.savefig('scatter_plot_with_Avg_HH_Tot_HH_regression.jpg')

```

***Observation:***

**In St. Lawrence, NY average size of household and total no of households have no correlation.**

***Interpretation:***

**1.The housing trends in this region does not follow any pattern where an increase or decrease in average household size corresponds to a proportional change in the total number of households in that region.**

**2. Employment opportunities in the area could be influencing the total number of households independently of the average size.**

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Example variables
variable1 = 'Total_UNITS_IN_1-unit_structures'
variable2 = 'Total_UNITS_IN_2-or-more-unit_structures'
variable3 = 'Total_UNITS_IN_Mobile_homes_and_all_other_types_of_units'

# Convert columns to numeric
merged_df[variable1] = pd.to_numeric(merged_df[variable1], errors='coerce')
merged_df[variable2] = pd.to_numeric(merged_df[variable2], errors='coerce')
merged_df[variable3] = pd.to_numeric(merged_df[variable3], errors='coerce')

# Calculate total counts for each variable
merged_df['total_1unit'] = merged_df[variable1]
merged_df['total_2unit'] = merged_df[variable2]
merged_df['total_mobile_and_other'] = merged_df[variable3]

# Calculate the total number of housing units for each geographic area
merged_df['total_units'] = merged_df[['total_1unit', 'total_2unit', 'total_mobile_and_other']].sum(axis=1)

# Calculate the percentage of each unit structure type
merged_df['percent_1unit'] = (merged_df['total_1unit'] / merged_df['total_units']) * 100
merged_df['percent_2unit'] = (merged_df['total_2unit'] / merged_df['total_units']) * 100
merged_df['percent_mobile_and_other'] = (merged_df['total_mobile_and_other'] / merged_df['total_units']) * 100

# Identify the dominant structure for each geographic area
merged_df['dominant_structure'] = merged_df[['percent_1unit', 'percent_2unit', 'percent_mobile_and_other']].idxmax(axis=1)

# Create a custom color map for each unit structure type
color_map = {
    'percent_1unit': 'red',
    'percent_2unit': 'green',
    'percent_mobile_and_other': 'blue'
}

# Map dominant structure to corresponding colors
merged_df['color'] = merged_df['dominant_structure'].map(color_map)

# Plot a map with color-coded dominant structures
fig, ax = plt.subplots(figsize=(9, 8))

# Plot each type separately with assigned colors
for structure_type, color in color_map.items():
    subset = merged_df[merged_df['dominant_structure'] == structure_type]
    subset.plot(ax=ax, color=color, linewidth=0.8, edgecolor='0.8', label=structure_type, aspect='equal')

# Set the title and axis properties
ax.set_title('Dominant Housing Structure by Geographic Area')
ax.set_axis_off()

# Create a custom legend
legend_elements = [Patch(facecolor=color, edgecolor='black', label=structure_type) for structure_type, color in color_map.items()]
ax.legend(handles=legend_elements, loc='upper left')

ax.set_axis_on()
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

plt.show()
```
